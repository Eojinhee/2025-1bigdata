```python
import requests
```

```python
url = "http://openapi.seoul.go.kr:8088/6757796d446a696e37394d5067586c/json/GetParkInfo/1/100/"
```

```python
response = requests.get(url)
```

```python
print(response.content)
```

```python
data = response.json()
```

```python
data
```

```python
import pandas as pd
```

```python
temp=[]

for n in data['GetParkInfo']['row']:
  temp.append(n)
temp
```

```python
pdData = pd.DataFrame(temp)
pdData
```

```python
import folium
folium.Map()
```

```python
anMap = folium.Map(location= [37.4965, 126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale=True)
anMap
```

```python
for n, row in pdData.iterrows():
  print(n, row['PKLT_NM'], row['LAT'], row['LOT'])
```

```python
for n, row in pdData.iterrows():
  mk = folium.Marker(location=[row['LAT'], row['LOT']], popup=row['PKLT_NM'])
  mk.add_to(anMap)
anMap
```

```python
anMap2 = folium.Map(location=[37.4965, 126.9573],
                    zoom_start=10,
                    zoom_control=False,
                    control_scale=True)
anMap2
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap2)

for n, row in pdData.iterrows():
  folium.Marker(location=[row['LAT'], row['LOT']],
                popup = folium.Popup(row['PKLT_NM'], max_width=200)
                ).add_to(mc)

anMap2
```
